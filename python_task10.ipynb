```python
# importing matplotlib library for plotting the graph
import matplotlib.pyplot as plt

#Importing csv module to use csv.reader object for reading the csv file
import csv

#creating two empty lists to populating the axes x,y forthe graph
x = []
y = []
z = []

#open method to open the file and return a file objet
file = open('company_sales_data.csv')

#csv readerobject to read the csv  file
csvreader = csv.reader(file)

# next method to obtain the first row(the header)
header = next(csvreader)

# iterate through the rows of the csvreader object and append  the empty lists for plotting the x- axix and y-axix
for row in csvreader:
    x.append(int(row[0]))
    y.append(int(row[2]))
    z.append(int(row[4]))
    
#closing the opened file 
file.close()

#plotting figure size  with user defined width and length
f = plt.figure()
f.set_figwidth(7)
f.set_figheight(6)

# plotting subplot using plt.subplot() of matplotlib library
# this subplot() of matplot lib is for the subplot sfor two rows
#In order to split the figure a 3-digit integer is given as a parameter to subplot().
#The integers describe the position of subplots. 
#first digit is the number of rows, the second is the number of columns, and the third is the index of the subplot.
#1st index of the subplots
plt.subplot(211)
plt.plot(x,z,color='k',linewidth=3,linestyle='-',marker='o',mfc='k')
plt.title('sale dataof a bathing soap')

#2nd index of the subplots
plt.subplot(212)
plt.plot(x,y,color='red', linewidth=3,linestyle='-',marker='o',mfc='red')
plt.title('sale data of a facewash')

#assigning names to the x-axis and y-axis of the plot
plt.xlabel('Month Number')
plt.ylabel('Sales units in  number')

#showing the plot
plt.show()

```
