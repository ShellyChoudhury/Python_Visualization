```python
# importing matplotlib library for plotting the graph
import matplotlib.pyplot as plt

#Importing csv module to use csv.reader object for reading the csv file
import csv

#creating 6 empty lists to populating the x,y axes forthe graph
x = []
y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
y6 = []

#open method to open the file and return a file objet
file = open('company_sales_data.csv')

#csv readerobject to read the csv  file
csvreader = csv.reader(file)

# next method to obtain the first row(the header)
header = next(csvreader)

# iterate through the rows of the csvreader object and append  the empty lists for plotting the x- axix and y-axix
for row in csvreader:
    x.append(int(row[0]))
    y1.append(int(row[1]))
    y2.append(int(row[2]))
    y3.append(int(row[3]))
    y4.append(int(row[4]))
    y5.append(int(row[5]))
    y6.append(int(row[6]))
#.close() method to close the opened file  
file.close()
#plotting figure size  with user defined width and length
f = plt.figure()
f.set_figwidth(7)
f.set_figheight(5)

# stackplot()function of the pyplot to plot the stacked area plot with  the required style properties 
plt.stackplot(x,y1,y2,y3,y4,y5,y6,colors =['purple','c','r','k','g','tab:olive'],labels = ['facecream','facewash','Tooth Paste','Bathing soap','Shampoo','Moisturiser'])


#displaying the legend inside the plot
plt.legend( loc= 'upper left')

#displaying the title of the plot
plt.title('all product sales data using stack plot')

#displaying the names of the x-axis and y-axis
plt.xlabel('Month Number')
plt.ylabel('sales unit in number')

#showing the plot
plt.show()
```
