```python
# importing matplotlib library for plotting the graph
import matplotlib.pyplot as plt

#Importing csv module to use csv.reader object for reading the csv file
import csv

#creating empty lists to populating the axes x,y forthe graph
x = []
y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
y6 = []

#open method to open the file and return a file objet
file = open('company_sales_data.csv')

#csv readerobject to read the csv  file
csvreader = csv.reader(file)

# next method to obtain the first row(the header)
header = next(csvreader)

# iterate through the rows of the csvreader object and append  the empty lists for plotting the x- axix and y-axix
for row in csvreader:
    x.append(int(row[0]))
    y1.append(int(row[1]))
    y2.append(int(row[2]))
    y3.append(int(row[3]))
    y4.append(int(row[4]))
    y5.append(int(row[5]))
    y6.append(int(row[6]))
#.close() method to close the opened file  
file.close()
#plotting figure size  with user defined width and length
f = plt.figure()
f.set_figwidth(7)
f.set_figheight(5)

#.plot()method to plot the lineplots with the required style properties 
plt.plot(x,y1,color='blue',linewidth=3,linestyle='-',marker='o',mfc='b',label= 'facecreame sales data')
plt.plot(x,y2,color='tab:orange',linewidth=3,linestyle='-',marker='o',mfc='tab:orange',label= 'facewash sales data')
plt.plot(x,y3,color='green',linewidth=3,linestyle='-',marker='o',mfc='g',label= 'toothpaste sales data')
plt.plot(x,y4,color='red',linewidth=3,linestyle='-',marker='o',mfc='r',label= 'bathing soap sales data')
plt.plot(x,y5,color='tab:purple',linewidth=3,linestyle='-',marker='o',mfc='tab:purple',label= 'shampoo sales data')
plt.plot(x,y6,color='tab:brown',linewidth=3,linestyle='-',marker='o',mfc='tab:brown',label= 'moisturizer sales data')


#displaying the legend inside the plot
plt.legend( loc= 'upper left')

#displaying th etitle of the plot
plt.title('sales data')

#displaying the name sof the x-axis and y-axis
plt.xlabel('Month Number')
plt.ylabel('sales unit in number')


#showing the plot
plt.show()
```
