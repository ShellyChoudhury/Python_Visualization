```python
# importing matplotlib library for plotting the graph
import matplotlib.pyplot as plt

#Importing csv module to use csv.reader object for reading the csv file
import csv

#creating two empty lists to populating the axes x,y forthe graph
x = []
y = []

#open method to open the file and return a file objet
file = open('company_sales_data.csv')

#csv readerobject to read the csv  file
csvreader = csv.reader(file)

# next method to obtain the first row(the header)
header = next(csvreader)

# iterate through the rows of the csvreader object and append  the empty lists for plotting the x- axix and y-axix
for row in csvreader:
    x.append(int(row[0]))
    y.append(int(row[3]))
    
print(x) 
print(y)  
    
#.close() method to close the opened file  
file.close()

#plotting figure size  with user defined width and length
f = plt.figure()
f.set_figwidth(7)
f.set_figheight(6)

# plotting scatter plot using plot() method of matplotlib library
plt.scatter(x,y,label = 'Tooth paste sales data')

# displaying the legend on theupper left position of the plot
plt.legend(loc= 'upper left')

# displaying gridinside the plot woth grid style
plt.grid(linestyle='--')

#assigning title of the plot
plt.title('toothpaste sales data')

#assigning names to the x-axis and y-axis of the plot
plt.xlabel('Month Number')
plt.ylabel('numbers of unit sold')

#showing the plot
plt.show()

```
