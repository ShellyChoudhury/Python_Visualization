```python
# importing matplotlib library for plotting the graph
import matplotlib.pyplot as plt

#Importing csv module to use csv.reader object for reading the csv file
import csv

#creating two empty lists to populating the axes x,y forthe graph
x1 = []
x2 = []
x3 = []
x4 = []
x5 = []
x6 = []

#open method to open the file and return a file objet
file = open('company_sales_data.csv')

#csv readerobject to read the csv  file
csvreader = csv.reader(file)

# next method to obtain the first row(the header)
header = next(csvreader)

# iterate through the rows of the csvreader object and append  the empty lists for plotting the x- axix and y-axix
for row in csvreader:
    x1.append(int(row[1]))
    x2.append(int(row[2]))
    x3.append(int(row[3]))
    x4.append(int(row[4]))
    x5.append(int(row[5]))
    x6.append(int(row[6]))
    

#.close() method to close the opened file  
file.close()

#adding all the values in each populated list to calculate the  total
total_x1 = sum(x1)
total_x2 = sum(x2)
total_x3 = sum(x3)
total_x4 = sum(x4)
total_x5 = sum(x5)
total_x6 = sum(x6)

# list of  all the  labels for each constituents  on the pie chart are stored in the labels variable
labels=['facecream','facewash','Toothpaste','bathingsoap','shampoo','moisturizer']

# list of respective colors in the colors variable
colors=['tab:blue','tab:orange','g','r','tab:purple','tab:brown']

#sizes variable stores the list of total vales of each list. This alos stores the data for plotting each constituent of the pie chart.
# all the variables have to ordered and matched up
sizes = [total_x1,total_x2,total_x3,total_x4,total_x5,total_x6]

#plt.pie()function to plot  the pie chart  
# startangle=90, makes the plotting right at the top middle of the pie chart. 
#The first elements are then plotted and the next sequential elements are plotted in a counter clockwise direction.
#To add percentages to each of the constitutents of the pie chart, I added , autopct='%1.1f%%', to the plt.pie() function.
#This formats the percentage to the tenth place.
plt.pie(sizes,labels=labels, colors=colors, startangle=90, autopct='%1.1f%%')

#displaying the legend inside the plot
plt.legend( loc= 'lower right')

# this line to get rid of the tilt in the pie chart that matplot creates by default
plt.axis('equal')

#displaying th etitle of the plot
plt.title('sales data')

#showing the plot
plt.show()
```
